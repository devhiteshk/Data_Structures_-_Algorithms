# Important Sorting Techniques

#### 1. Selection Sort




#### Working -->
```py

the list is divided into two parts, the sorted part at the left end and the unsorted part at the right end. Initially, the sorted part is empty and the unsorted part is the entire list.

The smallest element is selected from the unsorted array and swapped with the leftmost element, and that element becomes a part of the sorted array. This process continues moving unsorted array boundary by one element to the right.

This algorithm is not suitable for large data sets as its average and worst case complexities are of (n2), where n is the number of items.

Not Stable Sort

```

````python
def SelectionSort(arr):          # O(N^2) TC & O(1) SC --> NOT STABLE
    for i in range(len(arr)):

        min_ele = float('inf')
        min_ele_index = 0

        for j in range(i, len(arr)):
            curr_ele = arr[j]
            if curr_ele <= min_ele:
                min_ele = min(min_ele, curr_ele)
                min_ele_index = j

        arr[i], arr[min_ele_index] = arr[min_ele_index], arr[i]   
    return arr

arr = [4,2,-3,5,7,10,-1,12,9,13,22,100,0]
print(SelectionSort(arr))
````

#### 2. Bubble Sort 




#### Working -->

```py
Swap if element at i+1 index is greater than i, repeat this process n times, each time max val is at n-i index and at end array is sorted.

T.C. is O(N^2) and O(1) S.C. this is stable Sort
```

````python
def BubbleSort(arr):
    for i in range(len(arr)-1):
        for j in range(i+1, len(arr)):
            if arr[i]>arr[j]:
                arr[i], arr[j] = arr[j], arr[i]

    return arr

arr = [-3,100, -5,-1,32, 2, 3, 5, 6, 8, 23, 44, 77]
print(arr)
print(BubbleSort(arr))
````
